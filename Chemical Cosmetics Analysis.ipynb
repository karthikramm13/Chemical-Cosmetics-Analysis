### Analysis of Chemicals in Cosmetics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv("Chemicals in Cosmetics.csv")
dt.head()
```

```python
dt.isnull().sum()
```

```python
dt.shape
```

```python
dt.dtypes
```

```python
dt.shape
```

```python
dt=dt.drop(['DiscontinuedDate', 'ChemicalDateRemoved'], axis=1)
```

```python
dt.head()
```

```python
dt.dropna(axis='index',subset=['CSFId', 'BrandName', 'CasNumber'],inplace=True)
```

```python
dt.isnull().sum()
```

```python
dt['CSF'].fillna(method='ffill', inplace=True)
```

```python
dt.isnull().sum()
```

```python
dt['InitialDateReported']=pd.to_datetime(dt.InitialDateReported)
dt['MostRecentDateReported']=pd.to_datetime(dt.MostRecentDateReported)
dt['ChemicalCreatedAt']=pd.to_datetime(dt.ChemicalCreatedAt)
dt['ChemicalUpdatedAt']=pd.to_datetime(dt.ChemicalUpdatedAt)
dt.head(15)
```

```python
dt['ChemicalName'].unique()
```

```python
len(dt['ChemicalName'].unique())
```

### Companies with most Products

```python
plt.figure(figsize=(20,12))
dt['CompanyName'].value_counts().nlargest(15).plot.bar()
```

### Most used Chemicals

```python
import seaborn as sns
```

```python
Chemicals = dt['ChemicalName'].value_counts()[:5]
sns.barplot(Chemicals, Chemicals.index)
plt.xlabel('Count')
plt.title('Most Chemicals used')
```

### Top Primary Categories in Cosmetics

```python
plt.figure(figsize=(10,7))
prime_cat = dt['PrimaryCategory'].value_counts()[:5]
sns.barplot(x=prime_cat, y=prime_cat.index, palette='deep')
plt.title('Top Primary Categories in Cosmetics')
plt.xlabel('Number of Primary Categories')
```
